```python
import numpy as np
import matplotlib.pyplot as plt

# DÃ©finition des fonctions
def f1(x):
    return (1 - x**2) * x

def f2(x):
    return -np.sin(x)

def f3(x):
    return 1 - np.sin(x)

def f4(x):
    return 2 - np.sin(x)

```

```python
# ParamÃ¨tres
N = 100
xmin, xmax = -2, 2  # Intervalle conseillÃ© pour f1
Xmin, Xmax = -2*np.pi, 2*np.pi  # Intervalle conseillÃ© pour f2, f3, f4

# Subdivision de l'intervalle [xmin, xmax]
X = np.linspace(Xmin, Xmax, N)
X1 =  np.linspace(xmin, xmax, N)
# Calcul des valeurs de Y pour chaque fonction
Y1 = f1(X1)
Y2 = f2(X)
Y3 = f3(X)
Y4 = f4(X)

# TracÃ© des graphes
plt.figure(figsize=(10, 6))

plt.subplot(2, 2, 1)
plt.plot(X1, Y1)
plt.title("Graphe de f1(x)")
plt.xlabel("x")
plt.ylabel("f1(x)")

plt.subplot(2, 2, 2)
plt.plot(X, Y2)
plt.title("Graphe de f2(x)")
plt.xlabel("x")
plt.ylabel("f2(x)")

plt.subplot(2, 2, 3)
plt.plot(X, Y3)
plt.title("Graphe de f3(x)")
plt.xlabel("x")
plt.ylabel("f3(x)")

plt.subplot(2, 2, 4)
plt.plot(X, Y4)
plt.title("Graphe de f4(x)")
plt.xlabel("x")
plt.ylabel("f4(x)")

plt.tight_layout()
plt.show()

# Description qualitative des solutions
print("Description qualitative des solutions :")
print("1) Les solutions de f1(x) semblent osciller entre des valeurs positives et negatives.")
print("2) Les solutions de f2(x) oscillent sinusoidalement autour de zÃ©ro.")
print("3) Les solutions de f3(x) semblent osciller autour de 1 avec des variations sinusoÃ¯dales.")
print("4) Les solutions de f4(x) semblent osciller autour de 2 avec des variations sinusoÃ¯dales.")
```

```python
# Pour dÃ©crire de maniÃ¨re qualitative les solutions Ã  l'Ã©quation diffÃ©rentielle dx/dt = f(x)
# avec la condition initiale x(0) = x_init, nous allons examiner le comportement
# des solutions pour diffÃ©rentes valeurs de x_init dans l'intervalle conseillÃ©.

# 1) Pour f(x) = (1 - xÂ²)x sur l'intervalle [-2, 2]:
#    - Si x_init est proche de 0, les solutions oscillent autour de 0 avec une amplitude dÃ©croissante.
#    - Si |x_init| > 1, les solutions divergent vers l'infini lorsque t tend vers l'infini.
#    - Si |x_init| < 1, les solutions convergent vers 0 lorsque t tend vers l'infini.

# 2) Pour f(x) = -sin(x) sur l'intervalle [-2*pi, 2*pi]:
#    - Les solutions oscillent sinusoÃ¯dalement autour de 0 avec une pÃ©riode de 2*pi.
#    - Peu importe la valeur de x_init, les solutions oscillent autour de 0 et ne divergent pas.

# 3) Pour f(x) = 1 - sin(x) sur l'intervalle [-2*pi, 2*pi]:
#    - Les solutions oscillent sinusoÃ¯dalement autour de 1 avec une pÃ©riode de 2*pi.
#    - Peu importe la valeur de x_init, les solutions oscillent autour de 1 et ne divergent pas.

# 4) Pour f(x) = 2 - sin(x) sur l'intervalle [-2*pi, 2*pi]:
#    - Les solutions oscillent sinusoÃ¯dalement autour de 2 avec une pÃ©riode de 2*pi.
#    - Peu importe la valeur de x_init, les solutions oscillent autour de 2 et ne divergent pas.

# En rÃ©sumÃ©, pour chaque fonction f(x), les solutions oscillent ou convergent vers des valeurs
# spÃ©cifiques en fonction de la condition initiale x_init.
```

```python
#Exercice 2 - MÃ©thode dâEuler explicite
import numpy as np

def resoudre_euler(f, x_init, T_fin, N):
    # Initialisation des tableaux de temps et de solutions
    T = np.linspace(0, T_fin, N+1)
    X = np.zeros(N+1)
    X[0] = x_init
    dt = T_fin / N  # Pas de temps

    # Calcul des solutions par la mÃ©thode d'Euler explicite
    for i in range(N):
        X[i+1] = X[i] + dt * f(T[i], X[i])

    return T, X
```

```python
# Fonction f de l'Ã©quation diffÃ©rentielle
def f(t, x):
    return -x + np.sin(t)

# Condition initiale et paramÃ¨tres
x_init = 1
T_fin = 10
N = 100

# RÃ©solution de l'Ã©quation diffÃ©rentielle
T, X = resoudre_euler(f, x_init, T_fin, N)

# Affichage des rÃ©sultats
print("Temps (T) :\n", T)
print("Solutions (X): \n ", X)
```

```python
#Exercice 3 - Ãtude du pendule
x_init = 0
T_fin = 10

def pendule(x_init, T_fin, N):
    g=9.8
    l=1
    # Initialisation des tableaux de temps et de solutions
    T = np.linspace(0, T_fin, N+1)
    X = np.zeros(N+1)
    Y= np.zeros(N+1)
    X[0] = x_init
    Y[0]=0
    dt = T_fin / N  # Pas de temps

    # Calcul des solutions par la mÃ©thode d'Euler explicite
    for i in range(N):
        dXdt = Y[i]
        dYdt = - (g / l) * np.sin(X[i])
        
        # Mise Ã  jour des valeurs suivantes par Euler
        X[i+1] = X[i] + dt * dXdt
        Y[i+1] = Y[i] + dt * dYdt
    return T, X,Y

T, X, Y = pendule(0, T_fin, 1000)
plt.figure(figsize=(18, 10))
plt.subplot(2, 2, 1)
plt.plot(T, X)
plt.xlabel("T")
plt.ylabel("X")
plt.title("N=10000")

T, X, Y = pendule(1, T_fin, 1000)
plt.subplot(2, 2, 1)
plt.plot(T, X)
plt.xlabel("T")
plt.ylabel("X")
plt.title("N=10000")

T, X, Y = pendule(0.01, T_fin, 100000)
plt.subplot(2, 2, 1)
plt.plot(T, X)
plt.xlabel("T")
plt.ylabel("X")
plt.title("N=10000")

plt.tight_layout()
plt.show()
```

```python
import numpy as np

def pics_pendule(x_init, T_fin, N):
    g = 9.8  # IntensitÃ© de la pesanteur en m/s^2
    l = 1.0  # Longueur du pendule en mÃ¨tres
    
    # DÃ©finition du pas de temps
    dt = T_fin / N
    
    # Initialisation des tableaux pour stocker les rÃ©sultats
    T = np.linspace(0, T_fin, N+1)  # Tableau des temps
    X = np.zeros(N+1)               # Tableau des angles
    Y = np.zeros(N+1)               # Tableau des vitesses angulaires
    
    # Conditions initiales
    X[0] = x_init
    Y[0] = 0.0
    
    # Application de la mÃ©thode d'Euler explicite
    for i in range(N):
        # Calcul des dÃ©rivÃ©es aux temps actuels
        dXdt = Y[i]
        dYdt = - (g / l) * np.sin(X[i])
        
        # Mise Ã  jour des valeurs suivantes par Euler
        X[i+1] = X[i] + dt * dXdt
        Y[i+1] = Y[i] + dt * dYdt
    
    # Recherche des instants de temps oÃ¹ la vitesse angulaire change de signe (pics)
    pics_indices = []
    for i in range(N):
        if Y[i] > 0 and Y[i+1] <= 0 :
            pics_indices.append(i)
    
    # Calcul des instants de temps correspondants
    pics_temps = [T[i] for i in pics_indices]
    
    return pics_temps



# Exemple d'utilisation
x_init = np.pi / 4  # Angle initial en radians (45 degrÃ©s)
T_fin = 10.0         # Temps final en secondes
N = 1000            # Nombre de pas de temps

pics_temps = pics_pendule(x_init, T_fin, N)

# Affichage des rÃ©sultats
print("Liste des instants pour ",x_init," de temps oÃ¹ il y a un pic :", pics_temps)
print("La moyenne d'oscillation est ", np.mean(pics_temps))
print("")
x_init = np.pi / 6 # Angle initial en radians 
T_fin = 10.0         # Temps final en secondes
N = 1000            # Nombre de pas de temps

pics_temps = pics_pendule(x_init, T_fin, N)

# Affichage des rÃ©sultats
print("Liste des instants pour ",x_init," de temps oÃ¹ il y a un pic :", pics_temps)
print("La moyenne d'oscillation est ", np.mean(pics_temps))
print("")


x_init = np.pi / 2  # Angle initial en radians (45 degrÃ©s)
T_fin = 10.0         # Temps final en secondes
N = 1000            # Nombre de pas de temps

pics_temps = pics_pendule(x_init, T_fin, N)

# Affichage des rÃ©sultats
print("Liste des instants pour ",x_init," de temps oÃ¹ il y a un pic :", pics_temps)
print("La moyenne d'oscillation est ", np.mean(pics_temps))
print("")
```

```python
#Exercice 4 
# Pour une condition initiale x_init positive, l'Ã©quation logistique dÃ©crit la croissance de la population de poissons au fil du temps.
# L'Ã©quation diffÃ©rentielle dx/dt = x * (1 - x) dÃ©crit la variation de la population de poissons x par rapport au temps t.

# Lorsque x_init est positif, la population de poissons croÃ®tra initialement si x_init est infÃ©rieur Ã  1.
# Cependant, Ã  mesure que la population se rapproche de la capacitÃ© maximale de l'environnement (x = 1), la croissance diminuera progressivement jusqu'Ã  ce qu'elle se stabilise Ã  cette capacitÃ© maximale.

# Ainsi, pour x_init positif, le comportement de la population de poissons sera une croissance initiale suivie d'une stabilisation autour de la capacitÃ© maximale de l'environnement (x = 1).
```

```python
def h(x):
    return x*(1-x)

# Condition initiale et paramÃ¨tres
x_init = 1
T_fin = 20
N = 1000

# RÃ©solution de l'Ã©quation diffÃ©rentielle
T, X = resoudre_euler(h, x_init, T_fin, N)

# Affichage des rÃ©sultats

plt.figure(figsize=(18, 10))
plt.subplot(2, 2, 1)
plt.plot(T, X)
plt.xlabel("T")
plt.ylabel("X")
plt.title("N=10000")
```

```python
def resoudre_euler(h, x_init, T_fin, N):
    """
    RÃ©sout une Ã©quation diffÃ©rentielle avec la mÃ©thode d'Euler explicite.

    Arguments :
        h : fonction reprÃ©sentant la dÃ©rivÃ©e de la fonction Ã  rÃ©soudre
        x_init : condition initiale
        T_fin : temps final
        N : nombre de pas

    Retourne :
        T : tableau des temps
        X : tableau des solutions correspondantes
    """
    # Initialisation des tableaux de temps et de solutions
    T = np.linspace(0, T_fin, N+1)
    X = np.zeros(N+1)
    X[0] = x_init
    dt = T_fin / N  # Pas de temps

    # Calcul des solutions par la mÃ©thode d'Euler explicite
    for i in range(N):
        X[i+1] = X[i] + dt * h(X[i])

    return T, X

# ParamÃ¨tres
T_fin = 20
N = 200

# DÃ©finition de la fonction h
def f(x):
    return x * (1 - x)

# Affichage des courbes pour diffÃ©rentes conditions initiales
plt.figure(figsize=(10, 6))
for k in range(11):
    T, X = resoudre_euler(f, k/10, T_fin, N)
    plt.plot(T, X, label=f"x_init = {k/10}")

plt.xlabel("Temps")
plt.ylabel("Population de poissons")
plt.title("Ãvolution de la population de poissons avec diffÃ©rentes conditions initiales")
plt.legend()
plt.grid(True)
plt.show()
```
